```python
from transformers import AutoTokenizer,AutoModelForSequenceClassification
import torch
```

```python
tokenizer=AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
model=AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
sentence="I love this!"
```

```python
inputs=tokenizer(sentence,return_tensors="pt")
```

tokenizing the each word,the weight is also assigned to each word and padding also applied to the data then it was provided to output

```python
outputs=model(**inputs)
```

```python
prob=torch.softmax(outputs.logits,dim=1)
```

the raw data that is logits was passed as a parameter to activation function

```python
prob
```

```python
label_id=torch.argmax(prob).item()
label=model.config.id2label[label_id]
```

```python
label_id
```

```python
print(f"Sentence: {sentence}")
print(f"Sentiment: {label}")
print(f"probabilities: {prob.tolist()}")
```
